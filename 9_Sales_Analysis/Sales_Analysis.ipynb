```python
df['Date']=pd.to_datetime(df['Date'])
```

```python
df['Year']=df['Date'].dt.year
```

```python
df["Year_Month"]=df["Date"].dt.strftime('%Y-%m')
```

```python
df
```

```python
df.info()
```

```python
df['Profit']=df['Revenue'] - df['Cost']
```

```python
df
```

```python
df.info()
```

```python
df['Profit_Margins']=round(df['Profit']/df['Revenue'] *100,2)
```

```python
df.head()
```

```python
Month_Revenue=df.groupby(['Year_Month'])['Revenue'].sum().reset_index()
```

```python
Month_Revenue
```

```python
Month_Revenue=df.groupby(['Year_Month'])['Cost'].sum().reset_index()
```

```python
Grouped=df.groupby(['Year_Month'])[['Cost','Revenue','Profit']].sum().reset_index()
```

```python
Grouped
```

```python
import plotly.express as px
```

```python
fig=px.line(Grouped,x='Year_Month',y=['Cost','Revenue','Profit'],
            title='Monthly Performance')
fig.update_xaxes(title="Year-Month")
fig.update_yaxes(title="Amount($)")
fig.show()
```

```python
#group by sub category and calculate total quantity sold

Sucategory_Sales=df.groupby('Sub Category')['Quantity'].sum().reset_index().sort_values('Quantity',ascending=False)

```

```python
Sucategory_Sales
```

```python
Sucategory_Sales.sort_values('Quantity',ascending=False)
```

```python
fig=px.bar(Sucategory_Sales,y='Sub Category',x='Quantity',
           text_auto='.2s',title="Quantity by Sub Category")
fig.show()
```

```python
Sucategory_Profit=df.groupby('Sub Category')['Profit'].sum().reset_index()

Sucategory_Profit
```

```python
fig=px.bar(Sucategory_Profit,y='Sub Category',x='Profit',
           text_auto='.2s',title="Profit by Sub Category")
fig.update_xaxes(title='Profit($)')



fig.show()
```

```python
Sucategory_Profit_Margin=df.groupby('Sub Category')['Profit_Margins'].mean().reset_index()
Sucategory_Profit_Margin
```

```python
Sucategory_Profit_Margin=df.groupby('Sub Category')['Profit_Margins'].mean().round(2).reset_index()
Sucategory_Profit_Margin
```
