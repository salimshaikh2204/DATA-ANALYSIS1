predict the if a person would buy   life   insuarance or not 
based on his/her Age using logistic regression

```python
import pandas as pd
```

```python
import numpy as np
```

```python
from matplotlib import pyplot as plt
```

```python
%matplotlib inline
```

```python
df = pd.read_csv('insurance_data.csv')
```

```python
df.head()
```

```python
plt.scatter(df.age,df.bought_insurance,marker='*',color='red')
```

```python
from sklearn.model_selection import train_test_split     
```

```python
x_train,x_test,y_train,y_test = train_test_split(df[['age']],df.bought_insurance,train_size=0.8)
```

```python
x_train
```

```python
y_train
```

```python
x_test
```

```python
y_test
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
x_test
```

```python
y_predicted = model.predict(x_test)
```

```python
y_predicted
```

```python
model.coef_
```

```python
model.intercept_
```

```python
import math 
```

```python
def sigmoid(x):
    return  1 / ( 1+ math.exp(-x))
```

```python
def predict_function(age):
    z= model.coef_ * age  +  model.intercept_
    y=sigmoid(z)
    return y
```

```python
predict_function(55)
```

```python
predict_function(42)
```

```python
predict_function(25)
```
