diabetes patients classification using K-N-N ALgorithms

```python
import pandas as pd
import numpy as np
```

```python
from sklearn.model_selection import train_test_split

from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
data = pd.read_csv("diabetes.csv")
```

```python
data
```

```python
zero_not_accepted = ['Glucose','BloodPressure','SkinThickness','Insulin','BMI']
```

```python
for column in zero_not_accepted:
    data[column] = data[column].replace(0,np.nan)
    mean = int(data[column].mean(skipna=True))
    data[column] = data[column].replace(np.nan,mean)
```

```python
data.head()
```

```python
x = data.iloc[:,0:8]
```

```python
x
```

```python
y=data.iloc[:,8]
```

```python
y
```

split dataset

```python
x = data.iloc[:,0:8]
y=data.iloc[:,8]

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2)
```

```python

print(len(x_train))
print(len(y_train))
print(len(x_test))
print(len(y_test))
```

```python
classifier = KNeighborsClassifier(n_neighbors=11,p=2,metric='euclidean')
```

```python
sd = StandardScaler()
x_train =sd.fit_transform(x_train)
x_test = sd.fit_transform(x_test)
```

```python
x_train
```

```python
classifier.fit(x_train,y_train)
```

```python
y_pred =classifier.predict(x_test)
from sklearn.metrics import accuracy_score
y_pred
```

```python
x_train_prediction = classifier.predict(x_train)

training_data_accurancy = accuracy_score(x_train_prediction,y_train)
```

```python
print('Accurancy score of training data is ',training_data_accurancy)
```

```python
x_test_prediction = classifier.predict(x_test)

testing_data_accurancy = accuracy_score(x_test_prediction,y_test)
print('Accurancy score of testing data is ',testing_data_accurancy)
```
